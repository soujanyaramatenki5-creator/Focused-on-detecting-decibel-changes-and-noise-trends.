```python
# Import libraries
from gtts import gTTS
from IPython.display import Audio, display

# Take input from user
text = input("Enter text: ")

# Convert text to speech
tts = gTTS(text=text, lang="en")

# Save the audio file
tts.save("speech.mp3")

# Play the audio
display(Audio("speech.mp3"))
```
